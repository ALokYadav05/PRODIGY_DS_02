```python
plt.figure(figsize = (12,5))
sns.set_style('whitegrid')
ax = sns.countplot(data=train_df , x='Survived' ,palette='muted')

for bar in ax.patches:
  x = bar.get_x() + bar.get_width()/2
  y = bar.get_height()
  ax.annotate(
      text = f'{y}',
      xy = (x,y),
      xytext = (0,3),
      textcoords='offset points',
      ha = "center",
      va = "center",
      fontsize = 12,
      color = "black"

  )
ax.set_title("Total surviour v/s Non-survivour" , fontsize=12)
ax.set_xlabel(" Survived (0 = No , 1 = Yes) ", fontsize=12)
ax.set_ylabel(" Count ",fontsize=12)
plt.tight_layout()
plt.show()
```

```python
#insights : we can see that many people didnt survived
```

```python
plt.figure(figsize=(12,6))
sns.set_style("ticks")
ax = sns.countplot(data=train_df ,x='Sex',hue='Survived',palette='viridis')
for bar in ax.patches:
  x = bar.get_x() + bar.get_width()/2
  y = bar.get_height()
  ax.annotate(
      text = f'{y}',
      xy = (x,y),
      xytext = (0,3),
      textcoords='offset points',
      ha = "center",
      va = "center",
      fontsize = 12,
      color = "black"

  )

ax.set_title("Male v/s Feamle Survival rate" ,fontsize = 12 , fontweight='bold')
ax.set_xlabel("Sex (0 : Female , 1 : Male)" , fontsize = 12)
ax.set_ylabel(" Count ",fontsize= 12)
plt.show()
```

```python
# insights : female have more survival rates then male
```

```python
plt.figure(figsize=(12,6))
sns.set_style("darkgrid")
ax = sns.histplot(train_df['Age'] , bins=40 , kde=True , color = "red" , edgecolor="black" ,linewidth = 1.5)
ax.set_title(" Age-Distribution")
ax.set_xlabel("Age" , fontsize=12)
ax.set_ylabel("Count" , fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(train_df.corr(numeric_only=True), annot=True ,cmap='coolwarm',fmt='.2f' ,linewidth=0.5 ,linecolor="gray",square=True)
plt.title(" Correlation Heatmap of Numeric Feature" , fontsize=16 , fontweight="bold")
plt.xticks(rotation=45,ha="right" , fontsize=12)
plt.show()
```

```python
# Heatmap Insights:
# - Sex is strongly correlated with Survived (females more likely to survive)
# - Fare and Pclass show moderate correlation with Survived (higher fare & class = better survival)
# - Age, SibSp, and Parch have weak correlation with Survived
```
